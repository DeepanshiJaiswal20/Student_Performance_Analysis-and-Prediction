```python
from sklearn.linear_model import LinearRegression

score_data = df[["G1", "G2", "G3", "studytime", "failures", "absences"]]
predict = "G3"
x = np.array(score_data.drop([predict], 1))
y = np.array(score_data[predict])

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
linear_regression = LinearRegression()
linear_regression.fit(xtrain, ytrain)
```

```python
accuracy = linear_regression.score(xtest, ytest)
print(accuracy)
```

```python
predictions = linear_regression.predict(xtest)
for i in range(len(predictions)):
    print(predictions[x], xtest[x], [ytest[x]])
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(mean_squared_error(ytest, predictions))
print(r2_score(ytest, predictions))
```

```python
residual = ytest - predictions
sns.distplot(residual)
```
